# Linear Regression

```python
# Linear Regresion

age_car = [5,7,8,7,2,17,2,9,4,11,12,9,6]
speed_car = [99,86,87,88,111,86,103,87,94,78,77,85,86]

plt.scatter(age_car,speed_car)
plt.show()
```

```python
slope, intercept, r ,p, std_err = stats.linregress(age_car,speed_car)

def myfunc(x):
  return slope * x + intercept

# mymodel = list(map(myfunc,age_car))

# define by lambda or anonymous function in python

mymodel = list(map(lambda x: slope *x + intercept,age_car))

plt.scatter(age_car,speed_car)
plt.plot(age_car,mymodel)
plt.show()
```

```python
print(mymodel)
```

```python
print(f"r: {r}") # relationship between x and y (0 means no relationship)
print(f"p: {p}")
print(f"err: {std_err}")
```

```python
test_ages = [10,11,12,13,14,15]
test_values = list(map(myfunc,test_ages)) # insert testing value (age car = 10)
print(test_values)
```

```python
# adding new data to plot
plt.scatter(age_car,speed_car)

plt.plot(age_car,mymodel)
plt.scatter(test_ages,test_values,c='red')
plt.show()
```

# Polynomial Regression

solution of bad linear regression

```python
# looks imposible to fit in linear regression
x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]

plt.scatter(x, y)
plt.show()
```

polinomial

\[
f(x) = a_3x^3 + a_2x^2 + a_1x + a_0
\]

```python
# using polynomial

# 3 represent degree of polynomial
model = np.poly1d(np.polyfit(x,y,3))

# 1 is start point
# 22 is end point
line = np.linspace(1,22,100)

plt.scatter(x,y)
plt.plot(line,model(line))
plt.show()
```

```python
# evaluate the model
from sklearn.metrics import r2_score

print(f"r2_score = {r2_score(y,model(x))}")
```

```python
# testing the model
test_value = [2,10,11,4,14,22,20]
plt.scatter(x,y)
plt.plot(line,model(line))
plt.scatter(test_value,model(test_value))
plt.show()
```

```python
x = [89,43,36,36,95,10,66,34,38,20,26,29,48,64,6,5,36,66,72,40]
y = [21,46,3,35,67,95,53,72,58,10,26,34,90,33,38,20,56,2,47,15]

model = np.poly1d(np.polyfit(x,y,3))

line = np.linspace(2,95,100)

plt.scatter(x,y)
plt.plot(line,model(line))
plt.show()
```

```python
# evaluate the bad model

print(r2_score(y,model(x)))
```

# Multiple Regression

linear regression with more than one independent value

```python
import pandas as pd

df = pd.read_csv('data.csv')

df.head()
```

```python
X = df[['Volume', 'Weight']]
y = df.take([-1],axis=1) # CO2
```

```python
from sklearn import linear_model

regr = linear_model.LinearRegression()
regr.fit(X,y)
```

```python
#predict the CO2 emission of a car where the weight is 2300kg, and the volume is 1300cm3:
predicted = regr.predict([[2300,1300]])

print(f"predicted value = {predicted}")
```

```python
# important part
# the machine things that it has the coefisient
# so each value that we want to predict
# the target "y" will be increased by the multiplication of x and coef

print(f'coef = {regr.coef_}')
```

```python
# the coef is 0 [[0.00780526 0.00755095]]
# means that if we input [[3300,2300]]
# mathematicaly, it will be 107.xxx (from the predicted before) + (1000 * 0.00780526)
# 1000 = 3300 - 2300

# let's see

new_predict = regr.predict([[3300,1300]])
print(f'new predict {new_predict}')
```

# Scale

```python
from sklearn.preprocessing import StandardScaler

scale = StandardScaler()

df['Volume'] = df['Volume']/1000
df['Volume']

x = df[['Weight', 'Volume']]
y = df['CO2']

regr = linear_model.LinearRegression()
regr.fit(x,y)

# seems stupid
print(regr.coef_)
```

```python
# what if we put scaler so Weight and Volume has equal scale

scaledX = scale.fit_transform(x)

print(scaledX)
```

```python
# now create the model

model = linear_model.LinearRegression()
model.fit(scaledX,y)

print(model.coef_)
```

```python
# is it still same with before scaled?

# should use pipeline to prevent this act
scaled = scale.transform([[2300,3.3]])

# quite similiar
predicted = model.predict([scaled[0]])
print(f"predicted {predicted}")
```

```python
from sklearn.metrics import r2_score
print(r2_score(y,model.predict(scaledX)))
```
