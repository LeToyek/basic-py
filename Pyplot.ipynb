# Pie Charts

```python
x = np.array([35,25,15,55])

plt.pie(x)
plt.show()
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y,labels=mylabels)
plt.show()
```

```python
# explode for separate certain part
explode_list = np.zeros(len(y))
for i,val in enumerate(y):
  if val <= 20:
    explode_list[i] = 0.2

plt.pie(y,labels =mylabels,explode = explode_list, shadow = True)
plt.show()
```

```python
mycolors = ["black", "hotpink", "b", "#4CAF50"]

plt.pie(y,labels=mylabels,colors=mycolors)
plt.show()
```

```python
plt.pie(y,labels=mylabels)
plt.legend()
plt.show()
```
